```python
import torch
from tqdm.auto import tqdm
from pathlib import Path
from multiprocessing.pool import Pool, ThreadPool
```

```python
root = Path('/mnt/remote/data/users/thomasssajot/yolo_dataset/traffic_lights_2020/distorted/images/')
images = list(root.glob('**/*.jpeg'))
print(f'Found {len(images)} images.')
```

```python
labels_root = Path('/mnt/remote/data/users/thomasssajot/yolo_dataset/traffic_lights_2020/distorted/')
cross_prod_labels_root = labels_root / 'labels_cross_product_with_relevance'
coco_labels_root = labels_root / 'labels_coco'
```

```python
def convert_to_coco_label(file):
    with file.open('r') as f:
        new_labels = ''.join(['9 ' + line.split(' ', 1)[1] for line in f.readlines()])

    new_file = coco_labels_root / file.relative_to(cross_prod_labels_root)
    assert not new_file.stem.endswith('unixus')
    new_file = new_file.parent / (new_file.stem + 'unixus.txt')
    new_file.parent.mkdir(parents=True, exist_ok=True)
    with new_file.open('w') as f:
        f.write(new_labels)
```

```python

labels = list(cross_prod_labels_root.glob('**/*.txt'))

with Pool(24) as pool:
    list(tqdm(
        pool.imap(convert_to_coco_label, labels),
        desc='Generate coco labels where traffic light cls = 9',
        total=len(labels)
    ))
```
